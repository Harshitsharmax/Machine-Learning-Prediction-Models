**Problem Statement:**

This Calories_Burnt_Prediction model, employing XGBoost Regressor, forecasts calories burnt during exercise. It's trained on exercise datasets and calories datasets which are then concatinated, learning correlations between exercise features and calorie expenditure. After evaluation, it accurately predicts calorie burn based on exercise type, duration, and intensity, aiding in personalized fitness planning and monitoring.

```python
# Importing the Dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

Data Collection and Processing

```python
# Loading the data into a pandas DataFrame
calories = pd.read_csv('/content/calories.csv')
exercise = pd.read_csv('/content/exercise.csv')
```

```python
# Print first 5 rows of the calories DataFrame
calories.head()
```

```python
# Print first 5 rows of the exercise DataFrame
exercise.head()
```

Combining the two DataFrames

```python
calories_data = pd.concat([exercise,calories['Calories']],axis=1)
```

```python
calories_data.head()
```

```python
# Checking the number of rows and columns
calories_data.shape
```

```python
# Getting some information about the data
calories_data.info()
```

```python
# Checking the number of missing values in the data
calories_data.isnull().sum()
```

Converting the text data to numerical values in Gender Column

```python
calories_data.replace({"Gender":{'male':0,'female':1}}, inplace=True)
```

```python
calories_data.head()
```

Finding the Correlation in the Dataset

1. Positive Correlation
2. Negative Correlation

```python
correlation = calories_data.corr()
```

```python
# Constructing a heatmap to understand the correlation
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar = True, square = True, fmt = '.1f', annot = True, annot_kws={'size':18},cmap='Blues')
```

Separating the Features from the Targets

```python
x = calories_data.drop(columns=['Calories','User_ID'],axis=1)
y = calories_data['Calories']
```

```python
print(x)
```

```python
print(y)
```

Splitting the Data into training and testing data

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

Model Training

XGBoost Regressor

```python
# Loading the model
model = XGBRegressor()
```

```python
# Training the model with training data
model.fit(x_train,y_train)
```

Model Evaluation

```python
# Model accuracy on testing data
testing_prediction = model.predict(x_test)

# Mean Absolute Error
# It means the difference between true and predicted value
mae = metrics.mean_absolute_error(y_test,testing_prediction)

# R Squared Value
r2_test = metrics.r2_score(y_test,testing_prediction)
```

```python
print('Mean Absolute Error for testing data is =',mae)
print('r2_score for testing data is =',r2_test)
```
